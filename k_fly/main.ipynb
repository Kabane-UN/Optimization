```julia
using Plots
```

```julia
l_1(x) = (2*x-12)/3
h(x, y) = -2*x-3*y
x_1 = [0, 8]
g = [2, 3]*1.18
l_2(x) = 2-x
l_3(x) = (3*x-24)/-6
f(x) = (-2*x)/3
plot(x_1, [l_1(x) for x  x_1], color=:blue)
plot!(x_1, [l_2(x) for x  x_1], color=:blue)
plot!(x_1, [l_3(x) for x  x_1], color=:blue)
plot!(x_1, [0, 0], color=:darkviolet)
plot!([0, 0], [0, 5], color=:darkviolet)
plot!(x_1.+g[1], [f(x) for x  x_1].+g[2])
```

```julia
surface(0:1:7, 0:1:4, h, camera=(0, 30))
plot!(x_1, [l_1(x) for x  x_1], [0, 0], color=:blue)
plot!(x_1, [l_2(x) for x  x_1], [0, 0], color=:blue)
plot!(x_1, [l_3(x) for x  x_1], [0, 0], color=:blue)
plot!(x_1, [0, 0], [0, 0], color=:darkviolet)
plot!([0, 0], [0, 5], [0, 0], color=:darkviolet)
xlabel!("x")
ylabel!("y")
```
